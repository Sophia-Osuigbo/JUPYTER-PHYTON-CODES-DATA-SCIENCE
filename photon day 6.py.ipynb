```python
my_list = []
for number in range(0,500):
    if number % 2 == 0:
        my_list.append(number) 
print(my_list)
```

```python
my_list = []
for number in range(5,100):
    if number % 2 == 0:
        my_list.append(number)
        print(my_list)
```

```python
my_list = [ number for number in range(10,50) if number % 2 == 0]
print(my_list)
```
